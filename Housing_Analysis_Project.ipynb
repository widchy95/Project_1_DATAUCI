# Project Title: Housing Market in the U.S.

#### Project Description: An analysis of the United States Housing market aimed at identifying national trends and unique marketplace behaviors

As a result of the COVID-19 pandemic, many industries and markets experienced their own major paradigm shifts.  One of the most hard hitting and relevant sectors was the housing market.  Using datasets from Redfin, a major residential real estate brokerage, we conducted an analysis of the U.S. housing market to answer the following questions:

1. How have housing prices on the national level looked before, during, and after the COVID-19 pandemic (2019 - 2024))?
2. Which areas experienced the largest change in sales price and which experienced a decline?
3. Which markets could be ‘affordable’ and which are increasingly out of reach?
4. Which areas could be considered real estate ‘hotspots’ that people are looking to call home?


### 1. How have housing prices on the national and state level (before, during, and after the COVID-19 pandemic (2019 - 2024))?
Housing prices typically experience an increase year-over-year as a part of standard inflationary pressures from economic activity.  This is observed when looking at teh average median home prices at the national level (as illustrated by the bar chart showing the median price of homes sold):



However, following the COVID19 pandemic, a lot of the standard market pressures were disrupted and new behaviors emerged.  This manifested a sudden rise in housing prices in the years following the onset of the pandemic in 2019.  As per the following line graph of average median sales price, we can see that prices rose rapidply year of year as illustrated by the sharp change in slope following the year 2019.



Some of the influential factors to this rise in prices may include the rise of remote work, which fostered preferences for homes outside of cramped, expensive urban cities, supply chain disruptions, which led to shortages in construction materials and delays in building new homes, and fiscal policy, where government stimulus and lowered interest rates helped maintain or even grow consumer spending power, including purchases of new homes.  Interestingly, the stoppage of these policies, can be seen in the line graph in 2022 where the slope of the line suddenly slows down.  This behavior concides with the Federal Reserve increasing interest rates in an effort to curb inflation, signaling an end of the monetary government support experienced during the pandemic.


### 2. Which areas experienced the largest change in sales price and which experienced a decline?
It is important to note that the patterns seen in aggregate at the national level do not necessarily match what occurs locally.  On a state level, we see a more nuanced set of shifts.



Here we see that there are actually some states where homes are selling at prices that are on average lower than what they were just only a year ago.  The variations in YoY growth in median sale prices across different regions suggest a complex interplay of economic, demographic, and market factors at the local level that may be smoothed out when looking at the data nationally. Top regions by YoY growth typically benefit from strong economic performance, high demand, favorable policies, and limited supply. In contrast, bottom regions may suffer from economic challenges, population declines, oversupply, and less favorable market conditions.

### 3. Which markets could be ‘affordable’ and which are increasingly out of reach?
Amidst high levels of inflation, affordability has been top of mind, especially when it comes to housing.  Looking at the time frame of Pre-Covid19 (2017-2019) and Post-Covid19 (2020-June 2024), we see that the markets considered the most affordable and least affordable have shifted.

 

When considered what could be considererd the most affordable counties (based on Average Median Monthly Housing Price), we see that pre-COVID19, the least expensive counties were often in rural areas or regions with limited economic opportunities. These areas generally had lower property values as well as lower demand.  However, in the post-COVID19 era, some rural or previously less desirable areas are experiencing increased interest, largely due to their affordability and the location flexibility that remote work offers. This could lead to rising prices in those areas, potentially impacting the rankings of least expensive counties.  However, despite the rankings, the overal price levels across all the new 10 ten counties are much higher than the top counties in the pre-COVID19 era.


 

We see similar price increases when looking at the most expensive counties.  This is especially apparent in Nantucket county, where the average monthly sale price for a home is nearly double what it was prior to the pandemic.  There was also a significant increase in prices for areas considered non-major counties, such as Teton County in Wyoming.

### 4. Which areas could be considered real estate ‘hotspots’ that people are looking to call home?
Given the challenges in the housing market, where then are the in-demand locations that people are deciding is the next best place to call home?  Identifying these 'hotspots' in the housing market can be approached and interpreted in different ways.

For example, it may be plausible to consider a hot market as one where the most home are sold as illustrated in the charts below:

Major cities have high population densities, leading to a naturally larger pool of potential homebuyers, prior to COVID19. High demand for housing in these areas, driven by the concentration of jobs, amenities, and services, results in a higher volume of home sales.  Interestingly enough, after the pandemic hit, while there was a shift toward suburban and rural areas, major cities still maintained significant home sales, probably due to their large populations and ongoing demand from people who prefer urban living or cannot relocate.

 

Another possible metric to judge the rise or fall of a in-demand market could be to analyze the percentage change in home sales.  This approach can eliminate the bias/advantage that larger urban areas with large populations have when looking at raw home sales.



Through this lens, we still see that counties with the largest percentage increase in home sales is still occuring smaller non-urban counties, particularly in areas in the South and Midwest regions of the United States.


A third possible metric would be to analyze the 'Days on Market' which is defined as the duration between when a property is first put up for sale and when it goes under contract.  This metric serves as measure of motivation with the buyer-seller market.

 

Prior to COVID19, the fastest selling homes were found in more urban counties, with the fastest selling homes being on the market for about 2 weeks.  This changed dramatically following the pandemic, where now each of the top 10 fastest selling counties have homes put under contract within just 10 days of being put up for sale.  Again, the make-up of the counties pre-COVID19 vs. post-COVID19 illustrate and clear shift away for urban cities and towards non-urban suburban and rural communities.

### Conclusion

Prior to starting this analysis, our group hypothesized that Covid-19 had drastically impacted the housing market by reshaping general market behavior towards home ownership in major metropolitan cities. Despite the numerous factors that may impact this market, we narrowed down our parameters to analyze trends purely based on market changes. Based on our analysis, we saw the median price for homes in the US sharply rise between this period and saw a major uptick in sales volumes for counties outside of major metropolitan areas (though we do see states like Florida, California, and Texas experience rapid growth based on annual sales volume). We also noticed that several of the fastest growing markets post-COVID19 are in counties outside of major cities such as Los Angeles, New York, Atlanta, etc. and are instead growing in less densely populated counties like Cleveland County, NC or Beckham County, OK. We concluded that Covid-19 did create a major impact to the US housing market and the change in behavior between buyers and regional markets are clearly marked with a shift towards non-urban areas post-Covid. Of course, we acknowledge that there are still several factors outside of our analysis that does not capture the entire picture of general population behavior in the wake of Covid-19, the new market developments seen in our data do show a indication of a paradigm shift in housing desirability and costs.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import os
```

```python
# Load the data from the CSV file
data = pd.read_csv(r'data\data.csv', sep='\t')  # For tab-separated data
data_national = pd.read_csv(r'data\data_national.csv', sep='\t')  # For tab-separated data
data_national.head()
```

### Data Cleaning Phase

```python
# Check for missing values in the datasets
data.isnull().sum()
data_national.isnull().sum()
```

```python
# Fill missing values with the median value of each column for both datasets
data = data.fillna(data.median(numeric_only=True))
data_national = data_national.fillna(data_national.median(numeric_only=True))
```

```python

# Convert 'Month of Period End' to datetime format for both datasets
data['Month of Period End'] = pd.to_datetime(data['Month of Period End'], format='%B %Y')

data_national['Month of Period End'] = pd.to_datetime(data_national['Month of Period End'], format='%B %Y')

# Convert 'Homes Sold' and other numeric columns to numeric, coercing errors to NaN for both datasets
data['Homes Sold'] = pd.to_numeric(data['Homes Sold'].replace('[,]', '', regex=True), errors='coerce')
data['New Listings'] = pd.to_numeric(data['New Listings'].replace('[,]', '', regex=True), errors='coerce')
data['Inventory'] = pd.to_numeric(data['Inventory'].replace('[,]', '', regex=True), errors='coerce')
data['Days on Market'] = pd.to_numeric(data['Days on Market'].replace('[,]', '', regex=True), errors='coerce')

data_national['Homes Sold'] = pd.to_numeric(data_national['Homes Sold'].replace('[,]', '', regex=True), errors='coerce')
data_national['New Listings'] = pd.to_numeric(data_national['New Listings'].replace('[,]', '', regex=True), errors='coerce')
data_national['Inventory'] = pd.to_numeric(data_national['Inventory'].replace('[,]', '', regex=True), errors='coerce')
data_national['Days on Market'] = pd.to_numeric(data_national['Days on Market'].replace('[,]', '', regex=True), errors='coerce')

# Fill any NaN values in numeric columns with the median value of the respective columns for both datasets
data['Homes Sold'] = data['Homes Sold'].fillna(data['Homes Sold'].median())
data['New Listings'] = data['New Listings'].fillna(data['New Listings'].median())
data['Inventory'] = data['Inventory'].fillna(data['Inventory'].median())
data['Days on Market'] = data['Days on Market'].fillna(data['Days on Market'].median())

data_national['Homes Sold'] = data_national['Homes Sold'].fillna(data_national['Homes Sold'].median())
data_national['New Listings'] = data_national['New Listings'].fillna(data_national['New Listings'].median())
data_national['Inventory'] = data_national['Inventory'].fillna(data_national['Inventory'].median())
data_national['Days on Market'] = data_national['Days on Market'].fillna(data_national['Days on Market'].median())

# Remove currency symbols and convert 'Median Sale Price' to numeric for both datasets
data['Median Sale Price'] = data['Median Sale Price'].replace('[\$,K]', '', regex=True).astype(float) * 1000

data_national['Median Sale Price'] = data_national['Median Sale Price'].replace('[\$,K]', '', regex=True).astype(float) * 1000

# Group by quarter and calculate the median sale price for each quarter
quarterly_data = data.set_index('Month of Period End').resample('QE')['Median Sale Price'].median().reset_index()

quarterly_data_national = data_national.set_index('Month of Period End').resample('QE')['Median Sale Price'].median().reset_index()

# Create a new column for the quarter labels for both datasets
quarterly_data['Quarter'] = quarterly_data['Month of Period End'].dt.to_period('Q').astype(str)

quarterly_data_national['Quarter'] = quarterly_data_national['Month of Period End'].dt.to_period('Q').astype(str)

# Split the data into pre-COVID (before 2020) and post-COVID (from 2020 onwards) for both datasets
pre_covid_data = data[data['Month of Period End'] < '2020-01-01']
post_covid_data = data[data['Month of Period End'] >= '2020-01-01']

pre_covid_data_national = data_national[data_national['Month of Period End'] < '2020-01-01']
post_covid_data_national = data_national[data_national['Month of Period End'] >= '2020-01-01']

# Get unique months and corresponding median sale prices for pre-COVID 
pre_covid_months_national = pre_covid_data_national['Month of Period End'].dt.strftime('%Y-%m').unique()
pre_covid_median_prices_national = []
for date in pre_covid_months_national:
    pre_covid_median_prices_national.append(pre_covid_data_national[pre_covid_data_national['Month of Period End'] == date]['Median Sale Price'].median())

# Get unique months and corresponding median sale prices for post-COVID 
post_covid_months_national = post_covid_data_national['Month of Period End'].dt.strftime('%Y-%m').unique()
post_covid_median_prices_national = []
for date in post_covid_months_national:
    post_covid_median_prices_national.append(post_covid_data_national[post_covid_data_national['Month of Period End'] == date]['Median Sale Price'].median())
```

```python
# Look at housing price fluctuations during and following COVID-19 (2019 - 2024)
# Filter the dataset for the specified time frame and analyze price trends
covid_period = data[(data['Month of Period End'].dt.year >= 2017) & (data['Month of Period End'].dt.year <= 2024)]
price_trends = covid_period.groupby(covid_period['Month of Period End'].dt.year)['Median Sale Price'].mean()
price_trends
```

```python
# Plot topline national summary stats as a bar chart (Quarterly) -- National Dataset

# Split the data into pre-COVID and post-COVID
pre_covid_quarters = quarterly_data_national[quarterly_data_national['Month of Period End'] < '2020-01-01']
post_covid_quarters = quarterly_data_national[quarterly_data_national['Month of Period End'] >= '2020-01-01']

# Plot topline national summary stats as a bar chart (Quarterly) -- National Dataset
plt.figure(figsize=(20, 6))

# Plot pre-COVID data
plt.bar(pre_covid_quarters['Quarter'], pre_covid_quarters['Median Sale Price'], color='skyblue', label='Pre-COVID')

# Plot post-COVID data
plt.bar(post_covid_quarters['Quarter'], post_covid_quarters['Median Sale Price'], color='orange', label='Post-COVID')

# Highlight the start of COVID-19 with a vertical line
plt.axvline(x='2020Q1', color='red', linestyle='--', label='COVID-19 Onset')

# Customize the plot
plt.title('Topline National Summary Stats (Quarterly) - National')
plt.xlabel('Quarter')
plt.ylabel('Median Sale Price')
plt.xticks(rotation=90)
plt.legend()
plt.grid(axis='y')
plt.show()
```

### Question 2: Which areas experienced the largest change in sales and which experienced a decline?

```python
# Look at what states are experiencing the highest and lowest YoY growth in median sale prices
# Calculate Year-over-Year (YoY) growth in median sale prices
data['Year'] = data['Month of Period End'].dt.year
yoy_growth = data.groupby(['Region', 'Year'])['Median Sale Price'].mean().pct_change().unstack()

# Define number of top and bottom regions to display (adjust as needed)
num_top_regions = 10
num_bottom_regions = 10

# Get top and bottom regions for 2024
top_regions = yoy_growth[2024].nlargest(num_top_regions)
bottom_regions = yoy_growth[2024].nsmallest(num_bottom_regions)

# Combine top and bottom regions into a single DataFrame (optional for customization)
combined_regions = pd.concat([top_regions, bottom_regions])

combined_df = pd.concat([top_regions, bottom_regions], axis=1)
combined_df.columns = ['Top Growth', 'Bottom Growth']
combined_df
```

```python
# Bar chart to illustrate state-level YoY Growth in Median Sale Prices
# Create a color list for visual distinction
colors = ['skyblue'] * num_top_regions + ['orange'] * num_bottom_regions

# Create the combined bar chart
plt.figure(figsize=(12, 6))
combined_regions.plot(kind='bar', color=colors)

# Customize the chart title and labels 
plt.title('Top and Bottom Regions by YoY Growth in Median Sale Prices (2024)')
plt.xlabel('Region')
plt.ylabel('YoY Growth (%)')
plt.xticks(rotation=90)
plt.grid(True)
plt.legend(title='YoY Growth Category', labels=['Top Growth', 'Least Growth']) 
plt.show()
```
